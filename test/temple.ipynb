```python
import csv
import os
import re
from bs4 import BeautifulSoup
from urllib.request import urlopen, urljoin
import requests

url_o = "https://www.koreatemplefood.com/templefood/recipe/index.html?page="

n = 0
data = []  #    .

for i in range(1, 186):
    url = url_o + str(i)
    html = urlopen(url)

    soup = BeautifulSoup(html, 'html.parser')
    img_tags = soup.find_all('span', class_='img_src')
    
    for img_tag in img_tags:
        try:
            style_attr = img_tag['style']
            img_url_aa = re.search(r"url\('(.+?)'\)", style_attr)
            
            if img_url_aa:
                img_url = img_url_aa.group(1)
                img_url = urljoin("https://www.koreatemplefood.com", img_url)  # Make it an absolute URL
                
                # Extracting restaurant name
                restaurant_name = img_tag.find_next('p', class_='tit').get_text(strip=True)
                
                # Save data to the list
                data.append({'RestaurantName': restaurant_name, 'ImageURL': img_url})
                
                # Download and save image
                with requests.get(img_url, stream=True) as response:
                    response.raise_for_status()
                    img_extension = img_url.split('.')[-1]
                    img_file_path = f'./images/{restaurant_name}.{img_extension}'
                    with open(img_file_path, 'wb') as img_file:
                        for chunk in response.iter_content(chunk_size=8192):
                            if chunk:
                                img_file.write(chunk)
                                
                    print(f"Image {n} downloaded: {img_url}")
                    n += 1
        except Exception as e:
            print(f"Error downloading image {n}: {e}")

# Save data to CSV file
csv_file_path = 'restaurant_data.csv'
csv_columns = ['RestaurantName', 'ImageURL']

with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=csv_columns)
    writer.writeheader()
    for row in data:
        writer.writerow(row)

print("Download and CSV file creation complete.")
```
